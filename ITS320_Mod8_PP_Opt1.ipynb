```python
#------------------------------------------------------------------------
# Program Name: ITS320 Portfolio Project Module 8 Option 1
# Author: Devin Moya
# Date: 02/08/2020
#------------------------------------------------------------------------
#------------------------------------------------------------------------
# Program Inputs: Number for which task youd like to run, and Vehicle ID #'s
# Program Outputs: Autombile Inventory in console, and in text file, as well as some other prompts
#------------------------------------------------------------------------

class Automobile_Inventory:
    
    #Dictionary where all vehicle inventory will stored
    inventory_dict = {}

    #constructor w/ private attributes
    def __init__(self):
        self.__make = ""
        self.__model = ""
        self.__color = ""
        self.__year = 0
        self.__mileage = 0
        
    #adds new vehicle and attributes
    def add_vehicle(self):
        user_input = input("\nPlease enter a new vehicle inventory ID number: ")
        if user_input in Automobile_Inventory.inventory_dict.keys():
            print("\nID number already exists. Please enter another\n")
            self.add_vehicle()
        else:
            print("\nYou are entering info for new vehicle with ID:", user_input)
            self._attribute_values(user_input)
                                           
        
    #updates attributes of a vehicle in inventory
    def update_vehicle(self):
        user_input = input("\nEnter ID of vehicle you would like to update: ")
        if user_input in Automobile_Inventory.inventory_dict.keys():
            print("\nYou are updating info for vehicle with ID:", user_input)
            self._attribute_values(user_input)
        else:
            print("\nVehicle not found.\n")
            self.update_vehicle()
    
    #removes vehicle in inventory   
    def remove_vehicle(self):
        user_input = input("\nEnter ID of vehicle you would like to remove: ")
        if user_input in Automobile_Inventory.inventory_dict.keys():
            print("\nVehicle", user_input, " deleted.\n")
            del Automobile_Inventory.inventory_dict[user_input]
        else:
            print("\nVehicle not found.\n")
            self.remove_vehicle()
    
    #Private interface to input attributes for new vehicle or vehicle to update
    def _attribute_values(self, user_input):
        self.__make = input("\nPlease input make of vehicle. ")
        self.__model = input("Please input model of vehicle. ")
        self.__color = input("Please input color of vehicle. ")
        try:
            self.__year = int(input("Please input year of vehicle. "))
            self.__mileage = int(input("Please input mileage of vehicle. "))
        except:
            print("\nPlease enter a valid number. \n")
            self._attribute_values(user_input)
        
        Automobile_Inventory.inventory_dict[user_input] = {'Make': self.__make,
                                                            'Model': self.__model,
                                                            'Color': self.__color,
                                                            'Year': self.__year,
                                                            'Mileage': self.__mileage,
                                                           }
```

```python
menu_text = """
            1. View Vehicle Inventory
            2. Add Vehicle
            3. Update Vehicle
            4. Remove Vehicle
            5. Export inventory to text file
            6. Quit \n
            What would you like to do? """    
                

running = True
while running:
    decision = input(menu_text)
    instance = Automobile_Inventory()
    if decision == '1':
        if len(Automobile_Inventory.inventory_dict) > 0:
            for k,v in Automobile_Inventory.inventory_dict.items():
                print("ID:", k, "-", v)
        else:
            print("\nNo Vehicles in inventory.\n")
    elif decision == '2':
        instance.add_vehicle()
    elif decision == '3':
        instance.update_vehicle()
    elif decision == '4':
        instance.remove_vehicle()

    #Writes inventory to a .txt file
    elif decision == '5':
        i = 1
        f = open("Automobile_Inventory" + str(i) + ".txt","w")
        print("Writing Invertory to file: " + "Automobile_Inventory" + str(i) + ".txt")
        f.write(str(Automobile_Inventory.inventory_dict))
        f.close()
        print("Exporting complete.\n")
        i += 1
    elif decision == '6':
        print("\nGoodBye.\n")
        running = False
    else:
        print('Please choose a valid option.\n')
```
